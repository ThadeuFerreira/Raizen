```python
import pandas as pd
import xlrd 
```

```python
wb = xlrd.open_workbook('vendas-combustiveis-m3.xls')
```

```python
sheet_names = [name for name in wb.sheet_names() if name != 'Plan1' and name !='Sheet1']
```

```python
raw_data = pd.read_excel('vendas-combustiveis-m3.xls', sheet_name='Sheet1')
for sheet in sheet_names:
    raw_data = pd.concat([raw_data,pd.read_excel('vendas-combustiveis-m3.xls', sheet_name=sheet)], axis=0)
```

```python
df = pd.melt(raw_data, id_vars=['COMBUSTÍVEL', 'ANO', 'REGIÃO', 'ESTADO', 'UNIDADE',], value_vars=['Jan', 'Fev',
       'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'], var_name='MES', value_name='VOLUME')
```

```python
map_mes = {}
count = 1
for i in ['Jan', 'Fev',
       'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']:
    map_mes[i] = count
    count += 1

df['MES'] = df['MES'].map(map_mes)
```

```python
df['year_month'] =  df.agg('01/{0[MES]}/{0[ANO]}'.format, axis=1)
```

```python
df['year_month'] = pd.to_datetime(df['year_month'], format='%d/%m/%Y')
```
